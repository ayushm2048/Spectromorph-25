```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import soundfile as sf
from IPython.display import Audio

# Task 1

audio_path = 'audio.wav'
y, sr = librosa.load(audio_path, sr=None)

print(f"Sample rate : {sr} Hz")
print(f"Duration : {len(y)/sr:.2f} seconds")
Audio(y, rate=sr)
```

```python
# Task 2

plt.figure(figsize=(14, 4))
librosa.display.waveshow(y, sr=sr)
plt.title('Waveform')
plt.show()

D = librosa.stft(y)
S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)

plt.figure(figsize=(14, 5))
librosa.display.specshow(S_db, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar(format='%+2.0f dB')
plt.title('STFT Spectrogram')
plt.show()

plt.figure(figsize=(14, 5))
librosa.display.specshow(S_db, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Log-Scaled Spectrogram')
plt.show()
```

```python
# Task 3

# Zero Crossing Rate
zcr = librosa.feature.zero_crossing_rate(y)[0]
plt.figure(figsize=(14, 3))
plt.plot(zcr)
plt.title('Zero Crossing Rate')
plt.show()

# RMS Energy
rms = librosa.feature.rms(y=y)[0]
plt.figure(figsize=(14, 3))
plt.plot(rms)
plt.title('RMS Energy')
plt.show()

# Spectral Centroid
cent = librosa.feature.spectral_centroid(y=y, sr=sr)[0]
plt.figure(figsize=(14, 3))
plt.plot(cent)
plt.title('Spectral Centroid')
plt.show()

# Spectral Bandwidth
bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)[0]
plt.figure(figsize=(14, 3))
plt.plot(bw)
plt.title('Spectral Bandwidth')
plt.show()
```

```python
# Task 4
mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfcc, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
plt.show()
```

```python
# Task 5

# Time stretch
y_fast = librosa.effects.time_stretch(y, rate=1.5)

# Pitch shift
y_shift = librosa.effects.pitch_shift(y, sr=sr, n_steps=2)

print("Time stretched (1.5x) audio:")
display(Audio(y_fast, rate=sr))

print("Pitch shifted (+2 semitones) audio:")
display(Audio(y_shift, rate=sr))

sf.write('pitch_shifted.wav', y_shift, sr)
print("Pitch shifted audio saved")
```
